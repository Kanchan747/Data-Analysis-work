```python
# Assuming you have pandas and statsmodels installed
import pandas as pd
import statsmodels.formula.api as smf

# Read your data into a pandas DataFrame
data = pd.read_csv("DSC_503_mckinsey.csv")  # Assuming your data is in a CSV file

# Replace 'Sales_USD', 'Store_ID', and 'Store_Type' with your actual column names
model = smf.mixedlm("Sales_USD ~ Store_ID", data=data, groups=data["Store_Type"]).fit()
print(model.summary())
```

```python
# prompt: give me code to print VIF analysis and drop the columns containing string

import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load your data
data = pd.read_csv("DSC_503_mckinsey.csv")

# Drop the 'Sales_USD' column and string columns
data = data.select_dtypes(exclude=['object'])
data = data.drop('Sales_USD', axis=1)

# Calculate VIF for each numeric column
vif_data = pd.DataFrame()
vif_data["feature"] = data.columns
vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(len(data.columns))]

# Print the VIF values
vif_data

# Optional: Drop columns with high VIF (threshold > 10, adjust as needed)
# high_vif_columns = vif_data[vif_data['VIF'] > 10]['feature'].tolist()
# data = data.drop(high_vif_columns, axis=1)

# Print the updated DataFrame
# print(data.head())
```
