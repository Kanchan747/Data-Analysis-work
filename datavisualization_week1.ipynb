```python
import pandas as pd
df = pd.read_csv('NYBoroughs.csv')
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Borough', y='Population', data=df, palette='viridis')
plt.title('Population by Boroughs (2011)', fontsize=14, pad=10)
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```
